# Güneş Paneli Özelliklerinin DataFrame Olarak Elde Edilmesi (Deneme 2)

```python
features_lexron_solarpanel = []

for link in filtered_links:
    
    html = getAndParseURL(link)
    
    try:
        found_elements = html.find("p", class_="MsoNormal", style=lambda value: value and "MARGIN: 0cm 0cm 0pt; LINE-HEIGHT: normal").find_all_next('p')
        ürün_kodu = found_elements[3].text
    except:
        ürün_kodu= np.nan
    try:
        found_elements = html.find("p", class_="MsoNormal", style=lambda value: value and "MARGIN: 0cm 0cm 0pt; LINE-HEIGHT: normal").find_all_next('p')
        marka = found_elements[6].text
    except:
        marka = np.nan
    try:
        found_elements = html.find("p", class_="MsoNormal", style=lambda value: value and "MARGIN: 0cm 0cm 0pt; LINE-HEIGHT: normal").find_all_next('p')
        sertifika = found_elements[9].text
    except:
        sertifika= np.nan
    try:
        found_elements = html.find("p", class_="MsoNormal", style=lambda value: value and "MARGIN: 0cm 0cm 0pt; LINE-HEIGHT: normal").find_all_next('p')
        güç = found_elements[15].text
    except:
        güç = np.nan
    try:
        found_elements = html.find("p", class_="MsoNormal", style=lambda value: value and "MARGIN: 0cm 0cm 0pt; LINE-HEIGHT: normal").find_all_next('p')
        hücre_sayısı = found_elements[18].text
    except:
        hücre_sayısı = np.nan
    try:
        found_elements = html.find("p", class_="MsoNormal", style=lambda value: value and "MARGIN: 0cm 0cm 0pt; LINE-HEIGHT: normal").find_all_next('p')
        voc = found_elements[21].text
    except:
        voc = np.nan
    try:
        found_elements = html.find("p", class_="MsoNormal", style=lambda value: value and "MARGIN: 0cm 0cm 0pt; LINE-HEIGHT: normal").find_all_next('p')
        vmp = found_elements[24].text
    except:
        vmp = np.nan
    try:
        fiyat= html.find("div",{"class":"product-price-new"}).text
    except:
        fiyat=np.nan

    features_lexron_solarpanel.append([ürün_kodu,marka,sertifika,güç,hücre_sayısı,voc,vmp,fiyat])

df_lexron_colums=["Ürün Kodu","Marka","Sertifika","Güç","Hücre Sayısı","Açık Devre Voltajı(VOC)","Maksimum Voltaj(VMP)","Fiyat"]
df_lexron_solarpanel = pd.DataFrame.from_records(features_lexron_solarpanel, columns =df_lexron_colums) 


df_lexron_solarpanel.head(10)
```
