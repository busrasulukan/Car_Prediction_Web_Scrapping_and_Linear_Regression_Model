```python
# "sc-y5byif-0 kiHyCh" sınıfına sahip tüm div elementlerini seç
div_elements = soup.findAll("div", {"class": "sc-y5byif-0 kiHyCh"})

# Her bir div elementindeki <a> etiketlerinin href özelliklerini al
for div_element in div_elements:
    # <a> etiketlerini seç
    a_elements = div_element.find_all('a')
    
    # Her bir <a> etiketinin href özelliğini al
    href_list = [a.get('href') for a in a_elements]
    
    # Elde edilen linkler
    print(href_list)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup


# Web sürücüsünü başlat
driver = webdriver.Chrome()

# Her bir sayfa için işlemi tekrarla
href_list2 = []
for page in my_pages:
    # Sayfayı aç
    driver.get(page)
    
    # Sayfanın yüklenmesini bekleyerek 10 saniyeye kadar zaman tanı
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "sc-y5byif-0.kiHyCh"))
    )
    
    # Sayfanın kaynağını al
    page_source = driver.page_source
    
    # BeautifulSoup ile HTML içeriğini analiz et
    soup = BeautifulSoup(page_source, 'html.parser')
    
    # "sc-y5byif-0 kiHyCh" sınıfına sahip tüm div elementlerini seç
    div_elements = soup.select("div.sc-y5byif-0.kiHyCh")
    
    # Her bir div elementindeki <a> etiketlerinin href özelliklerini al
    for div_element in div_elements:
        # <a> etiketlerini seç
        a_elements = div_element.find_all('a')
        
        # Her bir <a> etiketinin href özelliğini al
        href_list2 = [a.get('href') for a in a_elements]
        
        # Elde edilen linkler
        print(href_list2)

# Web sürücüsünü kapat
driver.quit()
```

```python
base_url = "https://www.galaxus.de"

# Her bir linki base_url ile birleştir
links = [base_url + link for link in href_list2]

# Elde edilen linkler
print(links)
```

```python
features = []
for link in links:
    html = getAndParseURL(link)
    try:
        ürünadı = soup.find('div', {'class': "sc-uyg4v5-0 Cxrnw containerFluid"}).h1.text
    except:
        ürünadı = np.nan
    try:
        fiyat= soup.find('button', {'class': "sc-1aeovxo-5 jJOENO"}).span.text
    except:
        fiyat = np.nan
    try:
        uzunluk = soup.find('span', {'class': "sc-1omuje8-0 dixcWq"}).text
    except:
        uzunluk = np.nan
    try:
        genişlik = soup.find('span', {'class': "sc-1omuje8-0 dixcWq"}).findNext().findNext().text
    except:
        genişlik = np.nan
    try:
        hücretipi = soup.find('span', {'class': "sc-1omuje8-0 dixcWq"}).findAllNext()[9].text
    except:
        hücretipi = np.nan
    try:
        nominalgüç = soup.find('span', {'class': "sc-1omuje8-0 dixcWq"}).findAllNext()[15].text
    except:
        nominalgüç = np.nan
    try:
        paneltipi = soup.find('span', {'class': "sc-1omuje8-0 dixcWq"}).findAllNext()[21].text
    except:
        paneltipi = np.nan
    try:
        ürünkodu = soup.find('span', {'class': "sc-1omuje8-0 dixcWq"}).findAllNext()[27].text
    except:
        ürünkodu = np.nan
  
    features.append([ürünadı,fiyat,uzunluk,genişlik,hücretipi,nominalgüç,paneltipi,ürünkodu])
    time.sleep(1)
    
df_columns = ["Ürün Adı", "Fiyat","Uzunluk","Genişlik","Hücre Tipi","Nominal Güç","Panel Tipi","Ürün Kodu"]

df = pd.DataFrame.from_records(features, columns=df_columns)

df
```
